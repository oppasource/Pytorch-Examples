# Binary Classification using Pytorch

In this notebook we will create a dataset using scikit-learn for classification then we will build a simple 3 layer neural network and train it on this dataset. After training we will plot the decision boundary for the model to see how the model performs in order to classify datapoints.

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn.datasets

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

%matplotlib inline

# We will create a dummy dataset to play with and understand how pytorch can be used for this
n_samples = 400
x,Y = sklearn.datasets.make_moons(n_samples=n_samples, noise=0.3, random_state=1)
#x,Y = sklearn.datasets.make_circles(n_samples=n_samples, noise=0.25, random_state=1, factor=0.2)

plt.scatter(x[:,0], x[:,1], c=Y, cmap=plt.cm.Spectral)
plt.show()

# Creating one hot vector for target label
y = np.eye(3)[Y]
y = y[:,0:2]
```

```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

in_dim = 2
hidden_dim = 3
out_dim = 2
num_epochs = 1000
learning_rate = 0.1

# We are creating a 3 layer neural network architechture
# with 2 neurons in input layer, 3 neurons in hidden layer
# and 2 neurons in output layer so that one of them fires 
# up indicating that class input has been clasified to.
class Model(nn.Module):
    def __init__(self, in_dim, hidden_dim, out_dim):
        super(Model, self).__init__()
        self.linear1 = nn.Linear(in_dim, hidden_dim)
        self.sig1 = nn.Sigmoid()
        self.linear2 = nn.Linear(hidden_dim, out_dim)
        self.sig2 = nn.Sigmoid()

    def forward(self, inputs):
        out = self.linear1(inputs)
        out = self.sig1(out)
        out = self.linear2(out)
        out = self.sig2(out)
        return out
    

model = Model(in_dim, hidden_dim, out_dim)

# Binary Cross Entropy Loss
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate, betas=(0.9, 0.999))


# Training the Model 
for epoch in range(num_epochs):
    x_data = Variable(torch.from_numpy(x).float())
    y_data = Variable(torch.from_numpy(y).float())

    # Forward + Backward + Optimize
    optimizer.zero_grad()  
    outputs = model(x_data)
    loss = criterion(outputs, y_data)
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 100 == 0:
        print ('Epoch [%d/%d], Loss: %.4f' 
               %(epoch+1, num_epochs, loss.data[0]))
```

```python
#Plotting the decision boundary
h = 0.01
# create a mesh to plot in
x_min, x_max = x[:, 0].min() - 0.5, x[:, 0].max() + 0.5
y_min, y_max = x[:, 1].min() - 0.5, x[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Converting numpy array to torch tensors to feed in the model
test = np.c_[xx.ravel(), yy.ravel()]
test = Variable(torch.from_numpy(test).float())

# Getting output from the trained model 
Z = model(test).data.numpy()
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.scatter(x[:, 0], x[:, 1], c=Y, cmap=plt.cm.Reds)
plt.show()
```
